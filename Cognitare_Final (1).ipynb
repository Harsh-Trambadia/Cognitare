# Data Cleaning

```python
# Concatenet both dataframe to do cleaning simultaneously
```

```python
# add new column to identify both dataframe
train['test'] = 0
test['test'] = 1
```

```python
# concat both dataframe
total = pd.concat([train, test], axis=0)
```

```python
total.head()
```

```python
# Drop insignificant columns
total.drop(['NPD_WELL_BORE_CODE','NPD_WELL_BORE_NAME'], axis=1, inplace=True)
```

```python
total.head()
```

```python
# Check NULL values
total.isnull().sum()
```

```python
# Handel Null Values
```

```python
total['ON_STREAM_HRS'].fillna(total['ON_STREAM_HRS'].mean(), inplace=True)
```

```python
total['AVG_DOWNHOLE_PRESSURE'].fillna(train['AVG_DOWNHOLE_PRESSURE'].mean(), inplace=True)
total['AVG_DOWNHOLE_TEMPERATURE'].fillna(train['AVG_DOWNHOLE_TEMPERATURE'].mean(), inplace=True)
```

```python
total['AVG_DP_TUBING'].fillna(total['AVG_DP_TUBING'].mean(), inplace=True)
total['AVG_ANNULUS_PRESS'].fillna(total['AVG_ANNULUS_PRESS'].mean(), inplace=True)
total['AVG_CHOKE_SIZE_P'].fillna(total['AVG_CHOKE_SIZE_P'].mean(), inplace=True)
total['AVG_WHP_P'].fillna(total['AVG_WHP_P'].mean(), inplace=True)
total['AVG_WHT_P'].fillna(total['AVG_WHT_P'].mean(), inplace=True)
total['DP_CHOKE_SIZE'].fillna(total['DP_CHOKE_SIZE'].mean(), inplace=True)
total['BORE_GAS_VOL'].fillna(total['BORE_GAS_VOL'].mean(), inplace=True)
total['BORE_WAT_VOL'].fillna(total['BORE_WAT_VOL'].mean(), inplace=True)
```

```python
total.isnull().sum()
```

```python
# Make different dataframes for train and test data
```

```python
train_cleaned = total[total['test']==0]
test_cleaned = total[total['test']==1]
```

```python
# Drop extra column we added
train_cleaned.drop('test', axis=1, inplace=True)
test_cleaned.drop(['test','BORE_OIL_VOL'], axis=1, inplace=True)
```

```python
train_cleaned['BORE_OIL_VOL'].fillna(train_cleaned['BORE_OIL_VOL'].mean(), inplace=True)
```

# Data Visualization

```python
plt.figure(figsize=[10,10])
sns.heatmap(train_cleaned.corr(), annot=True, cmap='Blues')
plt.show()
```

```python

for i in train_cleaned.columns:
  if i != 'BORE_OIL_VOL':
    plt.figure(figsize=[8,8])
    sns.regplot('BORE_OIL_VOL', i, data=train_cleaned, scatter_kws={"alpha": 0.2})
    plt.xlabel('BORE_OIL_VOL')
    plt.ylabel(i)
    plt.title('Relation between Target and Features')
    plt.show()

```

# Model Building

```python
X = train_cleaned.drop('BORE_OIL_VOL', axis=1)
y = train_cleaned['BORE_OIL_VOL']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X,y, test_size=0.1)
```

```python
X_test = test_cleaned
```

1. Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
pred_lr = lr.predict(X_val)
```

2. Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor(max_depth=10)
```

```python
dt.fit(X_train, y_train)
```

```python
pred_dt = dt.predict(X_val)
```

3. Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=170, max_depth=10)
```

```python
rf.fit(X_train, y_train)
```

```python
pred_rf = rf.predict(X_val)
```

4. KNN

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn = KNeighborsRegressor()
```

```python
knn.fit(X_train, y_train)
```

```python
pred_knn = knn.predict(X_val)
```

5. XGBoost

```python
from xgboost import XGBRegressor
```

```python
xgb = XGBRegressor(max_depth=10)
```

```python
xgb.fit(X_train, y_train)
```

```python
pred_xgb = xgb.predict(X_val)
```

6. LightBGM

```python
from lightgbm import LGBMRegressor
```

```python
lgb = LGBMRegressor(max_depth=10)
```

```python
lgb.fit(X_train, y_train)
```

```python
pred_lgb = lgb.predict(X_val)
```

# Finding Accuracy

```python
from sklearn.metrics import r2_score
```

```python
acc_lr = r2_score(y_val, pred_lr)
acc_dt = r2_score(y_val, pred_dt)
acc_rf = r2_score(y_val, pred_rf)
acc_knn = r2_score(y_val, pred_knn)
acc_xgb = r2_score(y_val, pred_xgb)
acc_lgb = r2_score(y_val, pred_lgb)
```

```python
acc = [acc_lr, acc_dt, acc_rf, acc_knn, acc_xgb, acc_lgb]
```

```python
models = ['lr', 'dt', 'rf', 'knn', 'xgb', 'lgb']
```

```python
# Plot accuracy of each model to compare it
import plotly.graph_objects as go

fig = go.Figure(data=[go.Bar(
            x=models, y=acc,
            text=acc,
            textposition='auto'
        )])
fig.show()
```

```python
best_model = models[np.argmax(acc)]
```

```python
best_model
```
